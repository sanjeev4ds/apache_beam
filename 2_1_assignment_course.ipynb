```python
def regex_replace(record):
  return re.sub("[\[\]\-,!,.,?,:,',;]","",record)

def regex_split(record):
  return re.split(r"\t|\n| ", record)
```

```python
with beam.Pipeline() as p:
  input_collections = (
      p
      | 'read from text file' >> beam.io.ReadFromText('drive/MyDrive/APACHE_BEAM/datasets/data.txt')
      | 'replace characters' >> beam.Map(regex_replace)
      | 'split rows'>> beam.FlatMap(regex_split)
      | 'filter empty rows' >> beam.Filter(lambda record: record != '')
      | 'map 1 with reach record' >> beam.Map(lambda record: (record,1))
      | 'sum using combinekey'>> beam.CombinePerKey(sum)
      | beam.Map(print)
  )
```
